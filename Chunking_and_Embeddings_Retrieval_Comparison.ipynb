## Retrieval WITHOUT Chunking Cell

### Subtask:
Demonstrate information retrieval without any text chunking using FAISS and Sentence Transformers.


**Reasoning**:
The subtask requires demonstrating information retrieval without chunking. I will define queries, initialize embeddings, create a FAISS vector store from the full document, and then perform and display similarity search results for each query.



```python
queries = [
    "What is LangChain?",
    "How does LangChain help with reasoning and acting?",
    "What are the main components of LangChain?"
]

# Initialize SentenceTransformerEmbeddings
embeddings = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")
print("Embeddings model initialized.")

# Create a FAISS vector store directly from the document (no chunking)
# The 'document' object already exists from a previous step, so we wrap it in a list.
vector_store_no_chunking = FAISS.from_documents([document], embeddings)
print("FAISS vector store (no chunking) created.")

# Perform similarity search for each query and display results
for i, query in enumerate(queries):
    print(f"\n--- Retrieval Results (No Chunking) for Query {i+1}: {query} ---")
    retrieved_docs = vector_store_no_chunking.similarity_search(query, k=2)
    for j, doc in enumerate(retrieved_docs):
        print(f"Document {j+1} (score: {doc.metadata.get('score', 'N/A')}):\n{doc.page_content[:500]}...\n")

print("Information retrieval without chunking demonstrated successfully.")
```

**Reasoning**:
The previous step successfully demonstrated retrieval without chunking. The next logical step, according to the overall task of comparing different chunking strategies, is to implement and demonstrate fixed-length chunking.



```python
## Retrieval WITH Fixed-Length Chunking Cell

### Subtask:
Demonstrate information retrieval using fixed-length text chunking with `RecursiveCharacterTextSplitter`, FAISS, and Sentence Transformers.

#### Instructions:
1.  Initialize `RecursiveCharacterTextSplitter` with a `chunk_size` of 500 and a `chunk_overlap` of 50. Store it in a variable called `text_splitter`.
2.  Split the `sample_text` into chunks using `text_splitter.create_documents()`.
3.  Create a FAISS vector store named `vector_store_fixed_chunking` from the generated chunks and the existing `embeddings`.
4.  For each query in the `queries` list:
    a.  Perform a similarity search on `vector_store_fixed_chunking` using the query and retrieve the top 2-3 most relevant documents.
    b.  Print the query and the content of the retrieved documents, adding a header like 'Retrieval Results (Fixed-Length Chunking) for Query:' for each query.
```

**Reasoning**:
The previous code block contained markdown text within a python cell, leading to a `SyntaxError`. I need to correct this by translating the instructions into actual Python code to perform fixed-length chunking and retrieval.



```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

print("## Retrieval WITH Fixed-Length Chunking Cell")

# 1. Initialize RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
print("RecursiveCharacterTextSplitter initialized.")

# 2. Split the sample_text into chunks
# The 'document' object contains the full sample_text
fixed_length_chunks = text_splitter.create_documents([document.page_content])
print(f"Created {len(fixed_length_chunks)} fixed-length chunks.")

# 3. Create a FAISS vector store from the generated chunks
vector_store_fixed_chunking = FAISS.from_documents(fixed_length_chunks, embeddings)
print("FAISS vector store (fixed-length chunking) created.")

# 4. Perform similarity search for each query and display results
for i, query in enumerate(queries):
    print(f"\n--- Retrieval Results (Fixed-Length Chunking) for Query {i+1}: {query} ---")
    retrieved_docs = vector_store_fixed_chunking.similarity_search(query, k=2)
    for j, doc in enumerate(retrieved_docs):
        print(f"Document {j+1} (score: {doc.metadata.get('score', 'N/A')}):\n{doc.page_content[:500]}...\n")

print("Information retrieval with fixed-length chunking demonstrated successfully.")
```

## Overlapping Chunking Cell

### Subtask:
Generate a Code cell for retrieval using overlapping chunking. This cell will use a `RecursiveCharacterTextSplitter` with a chunk size of 500 characters and an overlap of 100 characters. **Reasoning**:
The subtask requires implementing overlapping chunking, creating a FAISS vector store with these chunks, and then performing retrieval for the defined queries. I will generate a code cell to execute all these steps, including printing the chunking parameters and retrieval results.



```python
print("## Retrieval WITH Overlapping Chunking Cell")

# 1. Initialize RecursiveCharacterTextSplitter with chunk_size=500 and chunk_overlap=100
chunk_size_overlap = 500
chunk_overlap_value = 100
text_splitter_overlap = RecursiveCharacterTextSplitter(chunk_size=chunk_size_overlap, chunk_overlap=chunk_overlap_value)

# 2. Print the chunk_size and chunk_overlap used
print(f"RecursiveCharacterTextSplitter initialized with chunk_size={chunk_size_overlap} and chunk_overlap={chunk_overlap_value}.")

# 3. Split the sample_text into chunks
# The 'document' object contains the full sample_text. create_documents expects a list of documents or strings.
overlapping_chunks = text_splitter_overlap.create_documents([document.page_content])

# 4. Print the total number of chunks created with overlapping
print(f"Created {len(overlapping_chunks)} overlapping chunks.")

# 5. Create a new FAISS vector store from the generated overlapping chunks
vector_store_overlapping_chunking = FAISS.from_documents(overlapping_chunks, embeddings)
print("FAISS vector store (overlapping chunking) created.")

# 6. Perform similarity search for each query and display results
for i, query in enumerate(queries):
    print(f"\n--- Retrieval Results (Overlapping Chunking) for Query {i+1}: {query} ---")
    retrieved_docs = vector_store_overlapping_chunking.similarity_search(query, k=2)
    for j, doc in enumerate(retrieved_docs):
        print(f"Document {j+1} (score: {doc.metadata.get('score', 'N/A')}):\n{doc.page_content[:500]}...\n")

print("Information retrieval with overlapping chunking demonstrated successfully.")
```
